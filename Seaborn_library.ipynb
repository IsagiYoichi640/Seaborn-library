```python
import numpy as np#mathematical operations
import matplotlib.pyplot as plt#basic graph plotting
import pandas as pd#dataset manipolation
import seaborn as sns #advanced graph plotting
```

# 1. Understanding the dataset

```python
ds=pd.read_csv('/content/USA_Housing.csv')
print("Dataset loaded")
```

```python
ds.head()
```

```python
#understand the colums to use
ds.info()
```

# 2. Data Visualization

```python
#understand the basic relationship found in the dataset
sns.pairplot(ds,palette="coolwarm")
#giving plot the title
plt.title("Basic Relationships Found In the dataset")
#Display the graph
plt.show()
```

```python
ds = ds.drop('Address', axis=1)
```

```python
plt.figure(figsize=(10,6))#height and width of heatmap
sns.heatmap(ds.corr(), annot=True, cmap="RdBu")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))#height and width of heatmap
sns.histplot(ds['Price'], kde=True, color="Purple")
plt.title("Distribution of House Prices")
plt.ylabel("Frequency")
plt.show()
```

```python
feature = ['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Area Population']
for feature in feature:
  plt.figure(figsize=(8,5))
  sns.scatterplot(data=ds,x=feature,y="Price",color='purple')
  plt.title(f"{feature} vs Price")
  plt.xlabel(feature)
  plt.ylabel("Price")
  plt.show()
```
